```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
lambda_rate = 10  # Average number of buses per hour
num_buses = 10000  # Total number of buses to simulate

# Simulate inter-arrival times (exponential distribution)
inter_arrival_times = np.random.exponential(scale=1/lambda_rate, size=num_buses)

# Simulate number of buses arriving in an hour (Poisson distribution)
hours = 1  # Time period to simulate buses arriving
bus_counts_per_hour = np.random.poisson(lam=lambda_rate * hours, size=num_buses)

# Plotting the inter-arrival times
plt.figure(figsize=(10, 6))
sns.histplot(inter_arrival_times, kde=True, bins=30, color="lightblue", stat="density")
plt.title("Distribution of Inter-Arrival Times (Exponential)")
plt.xlabel("Time Interval Between Buses (hours)")
plt.ylabel("Density")
plt.grid(True)
plt.show()

# Plotting the number of buses arriving per hour
plt.figure(figsize=(10, 6))
sns.histplot(bus_counts_per_hour, kde=False, bins=range(0, max(bus_counts_per_hour)+1), color="lightblue", stat="probability")
plt.title("Distribution of Buses Arriving in an Hour (Poisson)")
plt.xlabel("Number of Buses in an Hour")
plt.ylabel("Probability")
plt.grid(True)
plt.show()

# Summary statistics
print("Summary Statistics:")
print(f"Mean inter-arrival time: {np.mean(inter_arrival_times):.2f} hours")
print(f"Variance of inter-arrival time: {np.var(inter_arrival_times):.2f} hours^2")
print(f"Mean number of buses per hour: {np.mean(bus_counts_per_hour):.2f}")
print(f"Variance of buses per hour: {np.var(bus_counts_per_hour):.2f}")
```
